```python
import math

class BaseStatistics:
    def __init__(self,base_list):
        base_list.sort()
        self.ordered = base_list
        self.list_len = len(self.ordered)
        
    def calc_mean(self):
        if len(self.ordered) > 0:
            return round(sum(self.ordered)/self.list_len,2)

    def calc_weight_mean(self):
        if len(self.ordered) > 0:
            values = 0
            weight = [0.1,0.1,0.1,0.1,0.15,0.15,0.3]
            print(self.ordered)
            print(f"Original Mean:{round(sum(self.ordered)/self.list_len,2)}")
            
            for i in range(len(self.ordered)):
                values += weight[i]*self.ordered[i]
            print(f'The Weighted mean = {values}')

    def calc_median(self):
        if self.list_len % 2 == 0:
            pos = (self.list_len/2) - 1
            median = (self.ordered[int(pos+1)] + self.ordered[int(pos)])/2
        else:
            median = self.ordered[int(((self.list_len + 1)/2) - 1)] 
        return median
    
    def calc_mode(self):
        mode = None
        temp_mode = {}
        for num in self.ordered:
            if num not in temp_mode:
                temp_mode[num] = 0
            temp_mode[num] += 1
        # print(temp_mode)
        ans_sort = sorted(temp_mode.items(),key = lambda kv: kv[1] , reverse = True)
        # print(ans_sort)

        # Step 1 - Get the max number
        max_num = ans_sort[0][1]

        # Use the value to get all of the keys with that value using list comprehension
        if max_num > 1:
            mode = [num[0] for num in ans_sort if num[1] == max_num]
        return mode
        
    
    def calc_max(self):
        return max(self.ordered)
    
    def calc_min(self):
        return min(self.ordered)
    
    def calc_range(self):
        return self.calc_max() - self.calc_min()
    
    def calc_interval(self, divisor):
        res = None
        if divisor:
            res = self.calc_range()/divisor
        return res
    
    def calc_variance(self):
        mean = self.calc_mean()
        return sum([(x-mean)**2 for x in self.ordered])/(len(self.ordered)-1)
    
    def calc_stdev(self):
        return math.sqrt(self.calc_variance())
    
    
    def calc_skewness(self):
        mean = self.calc_mean()
        skew_part1 = sum([(x-mean)**3 for x in self.ordered])/(len(self.ordered)-1)
        return round(skew_part1/(self.calc_stdev()**3),3)
    
    
    def calc_adjusted_skewness(self):
        mean = self.calc_mean()
        skew_part1 = sum([(x-mean)**3 for x in self.ordered])/(len(self.ordered)-1)
        return round(math.sqrt((self.list_len*(self.list_len-1)))/(self.list_len-2),2) * (skew_part1/(self.calc_stdev()**3))
    
    def calc_excess_kurtosis(self):
        mean = self.calc_mean()
        kurtosis = sum([(x-mean)**4 for x in self.ordered])/(len(self.ordered)-1)
        return (round(kurtosis/(self.calc_stdev()**4),3))-3
    

    @staticmethod 
    def calc_covariance(list_x,list_y):
        mean_x = sum(list_x)/len(list_x)
        mean_y = sum(list_y)/len(list_y)
        x_1 = [(x-mean_x) for x in list_x]
        y_1 = [(y-mean_y) for y in list_y]
        return sum([x_1[i]*y_1[i] for i in range(len(x_1))])/(len(x_1)-1)
    

    @classmethod
    def calc_correlation(cls,list_x,list_y):
        '''
        We need to declare this as a class method to have access to the calc_stdev
        methods.  We have access using the cls argument.
        '''
        pass

# fire_flies_per_yard = [22,34,37,14,11,10]
# fire_flies = BaseStatistics(fire_flies_per_yard)
# print(' See FireFlies per yard BELOW \n','=='*35)  # Add a divider
# print(fire_flies.ordered)
# print(f"This is the mean of FireFlies per yard:  {fire_flies.calc_mean()}")
# print(f"This is the median of FireFlies per yard:  {fire_flies.calc_median()}")
# print()  # skip a line
# print(' See SUMMER TEMPERATURES BELOW \n','=='*35)  # Add a divider
# # # Do the same thing for temperatures
# summer_temps = [78, 77, 74, 80, 72]
# temps = BaseStatistics(summer_temps)
# print(temps.ordered)
# print(f"This is the mean of SUMMER TEMPERATURES per yard:  {temps.calc_mean()}")
# print(f"This is the median of SUMMER TEMPERATURES per yard:  {temps.calc_median()}")
print('='*35)
temperatures = [89, 93, 77, 79, 93, 95, 94]
icecream_sales = [20000, 25000, 16000, 18000, 21000, 24000, 23000]

g7 = BaseStatistics(temperatures)
g7.calc_weight_mean()
print('='*35)
t1 = BaseStatistics(temperatures)
stdev = t1.calc_stdev()
print(f'The standard deviation of temperatures = {stdev}')
interval = t1.calc_interval(4)
print(f'The interval size is = {interval}')
print('='*35)
c1 = BaseStatistics(icecream_sales)
stdev = c1.calc_stdev()
print(f'The standard deviation of icecream sales = {stdev}')
print('='*35)# skip a line
covar = BaseStatistics.calc_covariance(temperatures,icecream_sales)
print(f'The covariance = {covar}')
direction = 'positive' if covar > 0 else 'negative'
print(f'The covariance of temperatures and icecream sales is directionaly {direction}')
print('='*35)
correl = BaseStatistics.calc_correlation(temperatures,icecream_sales)
print(f'The correlation = {correl}')
signif = 'strong' if correl and abs(correl) >= .7 else 'weak'
print(f'The correlation of temperatures and icecream sales is {signif}')
print('='*35)

with open('grades.txt','r') as fh:
    line = fh.readline()
    # print(line)
    cntr = 0
    grades = []
    while line:
        # print(line)
        if cntr > 0:
            # skipped the header
            grade = line.strip('\n')
            grades.append(float(grade))
        line = fh.readline()
        cntr += 1
    # print(grades)
    
# g1 = BaseStatistics(grades)
# print(f'The mean = {g1.calc_mean()}')
# print(f'The median = {g1.calc_median()}')
# print(f'The mode = {g1.calc_mode()}')
# print(f'The max = {g1.calc_max()}')
# print(f'The min = {g1.calc_min()}')
# print(f'The range = {g1.calc_range()}')
print('='*35)

class_grades = {'Anton': [88, 78,83,85,87],
               'Rebecca': [92,97,95, 92, 88],
               'Franklin': [77, 78, 73,88, 90],
               'Orin': [88,78,98,92,85],
               'Rose': [90,92,94,96,98]}
# For each student
all_students = []
for k,v in class_grades.items():
    student = BaseStatistics(v)
    print(f'The mean for {k} = {student.calc_mean()}')
    print(f'The median for {k} = {student.calc_median()}')
    print(f'The mode for {k} = {student.calc_mode()}')
    print(f'The max for {k} = {student.calc_max()}')
    print(f'The min for {k} = {student.calc_min()}')
    print(f'The range for {k} = {student.calc_range()}') 
    print('='*40)
    all_students += v
# for the class
k = 'ALL STUDENTS'
print(all_students)
student = BaseStatistics(all_students)
print(f'The mean for {k} = {student.calc_mean()}')
print(f'The median for {k} = {student.calc_median()}')
print(f'The mode for {k} = {student.calc_mode()}')
print(f'The max for {k} = {student.calc_max()}')
print(f'The min for {k} = {student.calc_min()}')
print(f'The range for {k} = {student.calc_range()}') 
print('='*40)
```
