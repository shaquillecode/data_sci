stock_price = [23.5, 24.1, 22.7, 24.2, 25.2, 28.4, 27.8, 26.5, 27.5, 28.2]

<br>
1 - Calculate the mean stock price<br>
2 - Count the number of times the stock price was greater than 25<br>
3 - Calculate the daily rate of return for the series<br>
NOTE: rate of return is ((current price-previous price)/previous_price) * 100<br>
Example: price was 100 next day 105 then ((105-100)/100)*100 = 5% gain<br>
4 - How many days was the price up 3% or down 3% from the previous day?<br>

```python
stock_price = [23.5, 24.1, 22.7, 24.2, 25.2, 28.4, 27.8, 26.5, 27.5, 28.2]
# 1 - Calculate the mean stock price
def calc_mean(lst):
    return sum(lst)/len(lst) if len(lst) else None #Ternary Operator

print(f"1)The mean/Average is {round(calc_mean(stock_price),5)}")

# 2 - Count the number of times the stock price was greater than 25
counter = 0

for i in stock_price:
    # print(i) 
    if i > 25:
        counter += 1
print(f"2)All the stock prices greater than 25 are {counter}")

# 3 - Calculate the daily rate of return for the series
# Rate of return is ((current price-previous price)/previous_price) * 100
# Example: price was 100 next day 105 then ((105-100)/100)*100 = 5% gain


for i,k in enumerate(stock_price):
    print(i,k)

rate_of_return = []

for i in range(1, len(stock_price)):
    # ((current price - previous price)/previous_price) * 100
    rate_of_return.append(round(((stock_price[i] - stock_price[i-1])/stock_price[i-1])*100,2))
print(f"3){rate_of_return}")

# 4 - How many days was the price up 3% or down 3% from the previous day? 
counter2 = 0
for r in rate_of_return:
    if abs(r) > 3:
        counter2 += 1
print(f"4){counter2}")
```

Suppose we have data like the following:
sat_scores = [540, 590, 780, 720, 680, 570, 590, 600, 600, 610]
Calculate what percentage of students are above the 50th percentile and what percentage
score below the 50th percentile
HINT: You need to calculate the median of the data set

```python
sat_scores = [540, 590, 780, 720, 680, 570, 590, 600, 600, 610]

def calc_median(lst):
    ret_val = None
    lst.sort()
    cnt = len(lst)
    if cnt > 0:
        if cnt % 2 == 1:
            # odd number
            # middle number - for 3 items return index 1, for 5 items return index 2
            ret_val = lst[int((cnt -1)/2)]
        else:
            # Even number of items
            # 2 items add (x[0] + x[1])/2,
            # 4 items add (x[1] + x[2])/2,
            # 10 items add (x[4] + x[5])/2,
            val = int(len(lst)/2)
            # (x[val-1] + x[val])/2
            ret_val = (lst[val-1] + lst[val])/2
    return ret_val
sat_scores.sort()
print(sat_scores)
print(f"The median is {calc_median(sat_scores)}")

score_len = len(sat_scores)
if score_len % 2 == 0:
    #Even
    pos = int(score_len/2)
    median =(sat_scores[pos-1] + sat_scores[pos])/2
else:
    #Odd - middle
    pos = int((score_len-1)/2)
    median = sat_scores[pos]
print(f'Median = {median}')

# Percentage of students below the 50th percentile or Median
pct_below_median = (sum([1 for x in sat_scores if x < median])/score_len) * 100
print(f"{pct_below_median}% Below the Median")

# Percentage of students above the 50th percentile or Median
pct_above_median = (sum([1 for x in sat_scores if x >= median])/score_len)* 100
print(f"{pct_above_median}% Above the Median")







```
