```python
# strip vowels and constanants
import random

VOWELS = [i.lower() for i in 'AEIOUY']
class Words:
    def __init__(self,word, removal_char = '_'):
        self.word = word
        self.removal_char = removal_char
        self.current_state = self.random_remove()
        
        
    def count_vowels(self):
        return sum([1 for i in self.word if i in VOWELS] )
      
    def count_constanants(self):
        return sum([1 for i in self.word if not i in VOWELS] )
    
    def __str__(self):
        return(f'''The word is {self.word}.\n
        It has {len(self.word)} letters.\nVowel count = {self.count_vowels()}\n
        Constanants count = {self.count_constanants()}''')        
    
    def remove_vowels(self):
        return ''.join([self.removal_char if x in VOWELS else x for x in self.word])
    
    def remove_constanants(self):
        return ''.join([self.removal_char if not x in VOWELS else x for x in self.word])
    
    def strip_fixed(self,n=2):
        return ''.join([self.removal_char if x % n == 0 else self.word[x] for x in range(len(self.word))])

    def random_remove(self):
        options = list(range(len(self.word)))
        n = len(self.word)-1
        items = []
        for i in range(n,0,-1):
            x = random.choice(options)
            # print(f'You randomly choose {x}')
            items.append(x)
            # remove item from the list
        items = set(items)
        return ''.join([self.removal_char if i in items else char  for i,char in enumerate(self.word)])

    def missing_letter(self):
        # Including the number of times it is missing
        print('='*35)
        print(self.current_state)
        print(self.word)
        print('='*35)
    
word = Words('Characterization','_')
print(word)
print(f'Remove vowels = {word.remove_vowels()}')
print(f'Remove constanants = {word.remove_constanants()}')
print(f'Remove fixed = {word.strip_fixed(4)}')
print(word.random_remove())
print(f'The current state is {word.current_state}')

print(word.missing_letter())

phrase = 'I love coding in python'
wof = ''
for word in phrase.split():
    x = Words(word)
    nw = x.random_remove()
    wof += nw + '|' 
print(wof)
```

Implement a spin wheel method with the following:<br>
1 - Bankrupt<br>
2 - Lose turn<br>
3 - grand prize<br>
4 - Vacation<br>
5 - Car<br>
4 - Dollar amounts of 100 to 2000<br>
Every time you spin add the dollar amounts if bankrupt zero the amount<br>
and current player = False, 'Lose turn - keep money ' current_player = False<br>

```python
import random
DOLLAR  = [ (i*100) for i in range(1,21)]
PRIZES = ['Bankrupt', 'Lose Turn', 'Grand Prize', 'Vacation', 'Car'] + DOLLAR 

class Wheel_Of_Fortune:
    def __init__(self,player):
        self.player = player
        self.win_money = 0
        self.win_prize = []
    
    def spin_the_wheel(self):
        stop = False
        
        random_prize = random.choice(PRIZES)
        if random_prize in ['Bankrupt','Lose Turn']:
            self.win_money = 0 if random_prize == 'Bankrupt' else self.win_money
            self.win_prize.append(random_prize)
            stop = True
        elif random_prize in DOLLAR:
            print(f'You won {random_prize}')
            self.win_money += random_prize
        elif random_prize == 'Grand Prize' or random_prize ==  'Vacation' or random_prize == 'Car':
            self.win_prize.append(random_prize)
        print(f'Your prize is {self.win_money}/{self.win_prize}')
        return False if stop else True
        
    def __str__(self):
        return f'Player {self.player}.  Has {self.win_money}'
    
player1 = Wheel_Of_Fortune('Tom')
playing = True
while playing:
    playing = player1.spin_the_wheel()
    print(player1)
```
