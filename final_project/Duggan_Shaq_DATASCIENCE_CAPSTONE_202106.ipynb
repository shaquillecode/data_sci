```python
target_date = datetime.date(2021,10,12)

hiredate = df['hiredate'].unique()
hiredate.sort()
# print(hiredate)

latest_employed = max(df['hiredate'].unique())
earliest_employed = min(df['hiredate'].unique())

for hd in hiredate:
    df_male = df['hiredate'][df.gender == 'M'][df.hiredate == hd].count()
    df_female = df['hiredate'][df.gender == 'F'][df.hiredate == hd].count()
    
    y,m,d = latest_employed.split('-')
    y2,m2,d2 = earliest_employed.split('-')

    hired_date = datetime.date(int(y),int(m),int(d))
    hired_date2 = datetime.date(int(y2),int(m2),int(d2))

    latest_days = target_date - hired_date
    earliest_days = target_date - hired_date2
    
    yearsLate = int(str(latest_days)[:4]) * 0.002738  #==> Days to years
    yearsEarly = int(str(earliest_days)[:4]) * 0.002738 #==> Days to years

# print(df_male,df_female)
print(f" Shortest working employee was hired on {latest_employed} or {latest_days} \n which equals {round(yearsLate,1)} years")
print('==='*35)
print(f" Longest working employee was hired on {earliest_employed} or {earliest_days} \n which equals  {round(yearsEarly,1)} years")
```
