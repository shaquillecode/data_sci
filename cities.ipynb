<h2>Pandas - Remove a column</h2>
There are 2 ways to remove a column from a dataframe.<br>
<br>
Method 1<br
To drop a column use the drop method with the axis parameter
<br>
<code>
import pandas as pd
data = { 'New York':[20,15], 
    'California':[10,3], 
    'Rhode Island':[5,6], 
    'Texas': [15,3], 
    'California':[17,4], 
    'Florida':[8,9], 
    'New Hampshire':[3,4], 
    'New Jersey':[12,8]}
state_info = pd.DataFrame(data.values(),index = data.keys(),columns=['Yes','No'])
state_info['Total votes'] = state_info['Yes'] + state_info['No']
df = state_info.drop('Total votes', axis='columns')
df
</code>
NOTE: The drop is not executed in place.  You need to rename the dataframe to see<br>
    your changes

<h2>Exercise</h2>
Take the columns yes and no and transform them from values to a percentage of the votes for each State<br>
And make the data type a float32

```python
import pandas as pd
data = { 'New York':[20,15], 
    'California':[10,3], 
    'Rhode Island':[5,6], 
    'Texas': [15,3], 
    'California':[17,4], 
    'Florida':[8,9], 
    'New Hampshire':[3,4], 
    'New Jersey':[12,8]}
state_info = pd.DataFrame(data.values(),index = data.keys(),columns=['Yes','No']).astype('float32')
state_info['Total votes'] = state_info['Yes'] + state_info['No']
state_info['% Yes'] = ((state_info['Yes']/(state_info['Yes'] + state_info['No']))*100).astype('float64').round(2)
state_info['% No'] = ((state_info['No']/(state_info['Yes'] + state_info['No']))*100).astype('float64').round(2)
# state_info.pop('Total votes')
# state_info.pop('Yes')
# state_info.pop('No')
state_info
```

```python
import pandas as pd
import numpy as np
  
# creating and initializing a nested list
values_list = [[15, 2.5, 100], [20, 4.5, 50], [25, 5.2, 80],
               [45, 5.8, 48], [40, 6.3, 70], [41, 6.4, 90],
               [51, 2.3, 111]]
  
# creating a pandas dataframe
df = pd.DataFrame(values_list, columns=['Field_1', 'Field_2', 'Field_3'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
  
  
# Apply function numpy.square() to square 
# the values of 3 rows only i.e. with row
# index name 'a', 'e' and 'g' only
df = df.apply(lambda x: np.square(x) if x.name in [
              'a', 'e', 'g'] else x, axis=1)
  
# printing dataframe
df
```
