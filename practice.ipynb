The following represented the price of gasoline in the following states:
{'New York': [3.23, 3.17, 3.23, 3.10, 3.09],
'Connecticut': [3.15, 3.23, 3.11, 3.15, 3.09]}
Create a pandas series with the index of the state and the column with the average price
per 5 day period

```python
import pandas as pd

gasPrice5Day = {'New York': [3.23, 3.17, 3.23, 3.10, 3.09],
'Connecticut': [3.15, 3.23, 3.11, 3.15, 3.09]}

keys = gasPrice5Day.keys()
values = gasPrice5Day.values()

ps = pd.Series(values, index=keys)
nyGasPrices = ps.loc['New York']
ctGasPrices = ps.loc['Connecticut']
print(ps)
print(nyGasPrices, ctGasPrices)

#Find sum for NY list of 

sum1 = 0
i = 0
while(i < len(nyGasPrices)):
    sum1 += nyGasPrices[i]
    i = i + 1

print(f'The sum for 5 days in NY are: {sum1}')
print(f'The Average price per 5 day period for Gas Prices in New York are: {sum1/(len(nyGasPrices))}')

# Find Sum Using The Range Method for CT
sum2 = 0
for gas in range(0, len(ctGasPrices)):
    sum2 += ctGasPrices[gas]


print(f'The sum for 5 days in CT are: {sum2}')
print(f'The Average price per 5 day period for Gas Prices in Conneticut are: {sum2/(len(ctGasPrices))}')
```
