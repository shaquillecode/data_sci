<h2>Exercise</h2>
The following list represents the diameters of circles:<br>
<code>
circle_diameters = [1,2,3,5,8,13,21]
</code><br>
1 - Calculate the area of each circle<br>
2 - Calculate the circumference of each circle<br>
<br>
Use Numpy for the solution

<h4>Solution</h4>

```python
import numpy as np
import math
'''
Diameter: A line that passes through the center and its endpoints lie on the circle and represented by the letter 'd' or 'D'.

Radius: The distance from the center to a point on the boundary and represented by the letter 'r' or 'R'. 
 
Radius plays an important role in the formula for the area and circumference of a circle

Diameter formula: The diameter formula of a circle is twice its radius. Diameter = 2 Ã— Radius
d = 2r or D = 2R
'''

# The diameter of the circle is known
circle_diameters = [1,2,3,5,8,13,21]
# The radius can be calculated as: r = d/2 or R = D/2 
d = np.array(circle_diameters)
r = d/2
print(r)
# Area of a circle is pi * r ^2
circle_area = math.pi * (r ** 2)
print('Areas '+'='*35)
print(circle_area)
# Circumference of a circle is 2 * pi * r
circle_circumference = (2 * math.pi) * r
print('Circumferences '+'='*35)
print(circle_circumference)
```

<h2>Exercise</h2>
5 friends go out to dinner and their individual amounts were tallied on 1 bill<br>
<code>
dinner = [36, 42, 27, 32, 39]
</code><br>
Using numpy calculate the following:<br>
1 - The average meal price<br>
2 - The median meal price<br>
3 - The maximum cost for the meal<br>
4 - The least expensive meal<br>
5 - The difference each person varied from the mean<br>
6 - Add a 20% tip to each persons meal then Add 8.875 tax to the pre tax meal amount<br>
7 - What percentage of the total bill is each person contributing?<br>
8 - It was the birthday of the person who had the most expensive meal.  <br>
Distribute the cost of their meal amoungst the other patrons.<br>

<h4>Solution</h4>

```python
dinner = [36, 42, 27, 32, 39]
print()

sort_dinner = np.sort(dinner)
print(sort_dinner)
print('==='*35)

total_dinner = np.sum(sort_dinner)
guests = np.size(sort_dinner)

# 1 - The average meal price
mean = np.mean(sort_dinner)

# 2 - The median meal price
median = np.median(sort_dinner)

# 3 - The maximum cost for the meal
max_meal = np.max(sort_dinner)

# 4 - The least expensive meal
min_meal = np.min(sort_dinner)

# 5 - The difference each person varied from the mean
differ = [(x-mean) for x in sort_dinner]

# 6 - Add a 20% tip to each persons meal then Add 8.875 tax to the pre tax meal amount
tip_and_tax = [round(((0.20 * x)+x)+(0.0875 * x),2) for x in sort_dinner]

# 7 - What percentage of the total bill is each person contributing?
percent = [ f'{x} is contributing {round(((x/np.sum(tip_and_tax))*100),2)}% of total meal cost' for x in tip_and_tax]

print(F'The {guests} friends Average costs = {mean} \n Median = {median}\n Most expensive meal = {max_meal} \n Least expensive meal = {min_meal}')
print('==='*35)
print(F'Difference from the mean = {differ} \n The tip and tax included = {tip_and_tax}')
print('==='*35)
print(percent)

# 8 - It was the birthday of the person with the most expensive meal. 
# Distribute the cost of their meal amongst the other patrons

Birthday = [ f'{round(x+((np.max(tip_and_tax))/4),2)} is {round(((x/np.sum(tip_and_tax))*100),2)+((np.max(tip_and_tax))/4)}% of the Birthday meal costs' for x in tip_and_tax if x != np.max(tip_and_tax)]
Birthday
```

<h2>Casting</h2>
We can convert python data type into numpy data type using 2 methods<br>
<br>
Method 1<br>
Use dtype parameter<br>
<br>
<code>
arr_string = np.array(list(range(1,11)), dtype='S')
print(arr_string)
</code><br>
Method 2<br>
Use astype()<br>
<code>
import numpy as np
arr = np.array(list(range(1,11)), dtype='i8')
arr_2 = arr.astype('S')
print(arr_2)
</code>

```python
import numpy as np

arr_string = np.array(list(range(1,11)), dtype='S')
print(arr_string)

arr = np.array(list(range(1,11)), dtype='i8')
arr_2 = arr.astype('S')
print(arr_2)
```

<h2>Exercise</h2>
Take the following floating point number and cast them as integers<br>
Use both methods.<br>
<code>
rainfall = [2.3,3.7,2.4,1.9]
</code>

<h4>Solution</h4>

```python
import numpy

rainfall = [2.3,3.7,2.4,1.9]
arr = np.array(rainfall, dtype = 'f')
print(arr, arr.dtype)

arr = np.float32(rainfall)
print(arr, arr.dtype)
```
