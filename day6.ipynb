```python
import math

class Number_Theory(object):
    
    def __init__(self,number):
        self.number = number
        # self.prime_number = number
        
    def isPrime(self):
        prime = False
        if self.number > 1:
            prime = True
            for i in range(2,int(math.sqrt(self.number))+1):
                if self.number % i == 0:
                    prime = False
                    break
        return prime

    def square_root(self):
        return round(math.sqrt(self.number),4)
    
    def odd_or_even(self):
        return 'EVEN' if self.number % 2 == 0 else 'ODD'

    def factors(self):
        fact = []
        for i in range (1, int(self.number/2)+1):
            if (self.number % i == 0):
                fact.append(i)
        fact.append(self.number)
        return fact

    def prime_factors(self):
        return [x for x in self.factors() if self.isPrime(x)]
    
x = Number_Theory(7)  # instantiate with 7
print(x.isPrime())
print(x.square_root())
print(x.odd_or_even())
print(x.factors())
# print(x.prime_factors())
print('='*35)

x = Number_Theory(10)  # instantiate with 7
print(x.isPrime())
print(x.square_root())
print(x.odd_or_even())
print(x.factors())
# print(x.prime_factors())
```

```python
# Fix Code
import math
n = math.pi

class Shape:
    
    def __init__(self,name,dm1 = 0, dm2 = 0):
        self.name = name
        self.dm1 = dm1
        self.dm2 = dm2
        self.area = self.calc_area(dm1,dm2)
        self.perimeter = self.calc_perimeter(dm1,dm2)
        
    def calc_area(self, dm1, dm2):
        ret = None
        if self.name == 'square':
            ret = self.dm1**2
        elif self.name == 'rectangle':
            ret = self.dm1 * self.dm2
        elif self.name == 'circle': 
            ret = n * (self.dm1**2)
        return ret
    
    def calc_perimeter(self,dm1,dm2):
        res = None
        if self.name == 'square':
            res = 4 * self.dm1
        elif self.name == 'rectangle':
            res = (2 *self.dm1) + (2*self.dm2)
        elif self.name == 'circle': 
            res = 2 * n * self.dm1
        return res
    
    def __str__(self):
        return f'This shape is a {self.name} - It has an area of {self.area} and a perimeter of {self.perimeter}'
    


shape = Shape('rectangle',2,6)
print(shape)
# print(shape2.area)
# print(shape2.perimeter)

```

```python
# Open an account with Name and an initial deposit amount
# When you print the object have person deposit x amount and now has y amount on deposit

# Open an account
class openAnAccount:
    def __init__(self, bal = 1000):
        self.bal = 10000


    def withdraw(self):
        currentAmount=int(input('Enter the amount to withdraw:'))
        if(currentAmount > self.bal):
            print('Not enough funds to Withdraw!')
        else:
            self.bal = self.bal - currentAmount
            print('Current Balance:' , self.bal)


    def deposit(self):
        currentAmount=int(input('How much are you depositing:'))
        self.bal = self.bal + currentAmount
        print('The Balance you have now is = ',self.bal)
        
    
account= openAnAccount()
account.withdraw()
account.deposit()
```

```python
class Banking:
    custBalance = 0  # instance variable
    bankDeposits = []  # Class attribute
    
    def __init__(self,cust_name,amt=0):
        self.cust_name = cust_name
        self.amt = amt
        self.deposit(amt)

    def deposit(self, amt):
        self.custBalance += amt
        self.bankDeposits.append(amt)
                
    def __str__(self):
        return f'{self.cust_name} has {self.amt} on deposit'
    
cust1 = Banking('John',500)
print(cust1)

cust2 = Banking('Justine',1200)
print(cust2)

cust1.deposit(500)
print(f'{cust1.cust_name} has {cust1.custBalance} on deposit')

print(f'{cust2.cust_name} has {cust2.custBalance} on deposit')
cust2.deposit(10000)
print(f'{cust2.cust_name} has {cust2.custBalance} on deposit')
print(Banking.bankDeposits)
print(sum(Banking.bankDeposits))
```

```python
# Fix Code
def concat_objects(obj1, obj2):
    res = None
    if isinstance(obj1,int) and isinstance(obj2,int):
        res = obj1 * obj2
    elif isinstance(obj1,str) and isinstance(obj2,str):
        res = obj1 + obj2
    elif (isinstance(obj1,int) and isinstance(obj2,str)) or (isinstance(obj1,str) and isinstance(obj2,int)):
        res = obj1 * obj2
    elif isinstance(obj1,float) and isinstance(obj2,float):
        res = obj1 * obj2
    else:
        res = "Can't do Floats and Numbers. Sorry."
    return res

print(concat_objects('Shaq', 3))
print(concat_objects(2, 3))
print(concat_objects(3, 'Shaq'))
print(concat_objects('Shaq', 'loves coding in Python'))
print(concat_objects('Shaq', 3.2))
print(concat_objects(None, None))
```

```python
# Fix Code
# Class  method and static methods
# Python program to demonstrate
# use of class method and static method
from datetime import date

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        

# a class method to create a Person object by birth year.
    def fromBirthYear(cls,name, year):
        return cls(name, date.today().year - year)

# a static method to check if a person is adult or not.
    @staticmethod
    def isAdult(age):
        return age > 18

    def __str__(self):
        return f'{self.name} is {self.age} years old'

    

person1 = Person('Hanna', 32)
print(person1)
print('ADULT' if Person.isAdult(16) else 'MINOR')
```

in the following sequence:<br>
<code>
x = [1,2,0,5,1,4,-1,2]

```python
x = [1,2,0,5,1,4,-1,2]

def greatNumber(lst):
    theMax = lst[0]
    for num in lst:
        # print(num)
        if theMax < num:
            theMax = num
    return theMax
  
res = greatNumber(x)
print(res)

```
