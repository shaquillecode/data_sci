<h2>Exercise</h2>
The following list represents the diameters of circles:<br>
<code>
circle_diameters = [1,2,3,5,8,13,21]
</code><br>
1 - Calculate the area of each circle<br>
2 - Calculate the circumference of each circle<br>
<br>
Use Numpy for the solution

<h4>Solution</h4>

```python
import numpy as np
import math

# Area of a circle is 2 pi r
circle_diameters = [1,2,3,5,8,13,21]
circle_area = ((np.array(circle_diameters) /2) ** 2) * math.pi
print('Areas '+'='*35)
print(circle_area)
# circumference of a circle is 2 pi r
circle_circumference = np.array(circle_diameters) * math.pi
print('Circumferences '+'='*35)
print(circle_circumference)
```

```python
apartment = {'Bedroom 1':(12,12), 'Bedroom 2': (12,10), 'Bathroom 1': (6,8),
    'Bathroom 2': (6,8), 'Kitchen': (10,8), 'Foyer': (14,4), 'Dining Room': (12,10),
    'Living Room': (15,22)}
print('Area '+'='*35)
square_footage = [f'{k} - Total area = {v[0]*v[1]}' for k,v in apartment.items()]
print(square_footage)

print('Total Square feet '+'='*35)
tot_sq_feet = sum([v[0]*v[1] for k,v in apartment.items()])
print(f'The total square feet of the apartment is {tot_sq_feet}')

print('Total Square feet '+'='*35)
tot_sq_feet = sum([v[0]*v[1] for k,v in apartment.items()])
print(f'The apartment will sell for {tot_sq_feet*12*90}')
```

<h2>Casting</h2>
We can convert python data type into numpy data type using 2 methods<br>
<br>
Method 1<br>
Use dtype parameter<br>
<br>
<code>
arr_string = np.array(list(range(1,11)), dtype='S')
print(arr_string)
</code><br>
Method 2<br>
Use astype()<br>
<code>
import numpy as np
arr = np.array(list(range(1,11)), dtype='i8')
arr_2 = arr.astype('S')
print(arr_2)
</code>

```python
import numpy as np

arr_string = np.array(list(range(1,11)), dtype='S')
print(arr_string)

arr = np.array(list(range(1,11)), dtype='i8')
arr_2 = arr.astype('S')
print(arr_2)
```

<h2>Exercise</h2>
Take the following floating point number and cast them as integers<br>
Use both methods.<br>
<code>
rainfall = [2.3,3.7,2.4,1.9]
</code>

<h4>Solution</h4>

```python
import numpy

rainfall = [2.3,3.7,2.4,1.9]
arr = np.array(rainfall, dtype = 'f')
print(arr, arr.dtype)

arr = np.float32(rainfall)
print(arr, arr.dtype)
```
