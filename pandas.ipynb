<h2>Docstrings</h2>

```python
def reverse_string(phrase):
    '''
    This function takes in a phrase and prints the phrase in reverse order
    '''
    return ''.join([phrase[x] for x in range(len(x)-1,-1,-1)])

x = 'I love coding in python'
print(reverse_string(x))
print(help(reverse_string))
```

<h2>Adding to a data series</h2>
To add to a series we use the append method.  However, we must use an optional<br>
parameter if we haven't included an index on our series.<br>
<code>
import pandas as pd
x = range(101,121,2)
ps = pd.Series(x)
y = range(1,11)
p2 = pd.Series(y)
p3 = ps.append(p2, ignore_index=True)
p3
</code>

```python
import pandas as pd
x = range(101,121,2)
ps = pd.Series(x)
y = range(1,11)
p2 = pd.Series(y)
p3 = ps.append(p2, ignore_index=True)
p3
```

<h2>Removing a row from a series</h2>
To remove a line from a series use the drop method<br>
<code>
import pandas as pd
x = range(101,121,2)
ps = pd.Series(x)
y = range(1,11)
p2 = pd.Series(y)
p3 = ps.append(p2, ignore_index=True)
p4 = p3.drop(1)
p4
</code>

```python
import pandas as pd
x = range(101,121,2)
ps = pd.Series(x)
y = range(1,11)
p2 = pd.Series(y)
p3 = ps.append(p2, ignore_index=True)
p4 = p3.drop(1)
p4
```

<h2>Pandas Series Index</h2>

<code>

</code>
<br>

```python
import pandas as pd
phrase = ['A','B','C','D','E']
index = range(5,10)
ps = pd.Series(phrase, index = index)
ps
```
