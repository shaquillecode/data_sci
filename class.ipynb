<h2>Exercise</h2>
Create a class called shape.  Set a shape name in the constructor.<br>
Make sure your instance name is in lower case all the time.<br>
When you print the instance it prints the following 'This shape is a 'shape name'

```python
import math

class Shape:
    
    def __init__(self,name,dm1 = 0, dm2 = 0):
        self.name = name
        self.dm1 = dm1
        self.dm2 = dm2
        self.area = self.calc_area(dm1,dm2)
        self.perimeter = self.calc_perimeter(dm1,dm2)
        

    def calc_area(self, dm1, dm2):
        res = None
        if self.name == 'square':
            res =  self.dm1 **2
        elif self.name == 'rectangle':
            res = self.dm1 * self.dm2
        elif self.name == 'circle':
            res = round(3.14*(self.dm1**2),2)
        return res
    
    
    def calc_perimeter(self,dm1, dm2):
        res = None
        if self.name == 'square':
            res = 4*self.dm1
        elif self.name == 'rectangle':
            res = 2*self.dm1 + 2*self.dm2
        elif self.name == 'circle':
            res = round(2*3.14*(self.dm1),2)
        return res
    
    def __str__(self):
        return f'This shape is a {self.name} it has an area of {self.area} and a perimeter of {self.perimeter}'
    
shape1 = Shape('square',4)
print(shape1)
# print(shape1.area)
# print(shape1.perimeter)

shape2 = Shape('rectangle',2,6)
print(shape2)

shape3 = Shape('circle',3)
print(shape3)
```

<h2>Banking App Model</h2>

Open an account with Name and possibly an initial deposit amount
When you print the object have This person deposited X amount and now has Y amount on deposit


```python
class Banking:
    custBalance = 0  # instance variable
    bankDeposits = []  # Class attribute
    
    def __init__(self,cust_name,amt=0):
        self.cust_name = cust_name
        self.amt = amt
        self.deposit(amt)

    def deposit(self, amt):
        self.custBalance += amt
        self.bankDeposits.append(amt)
                
    def __str__(self):
        return f'{self.cust_name} has {self.amt} on deposit'
    
cust1 = Banking('John',500)
print(cust1)

cust2 = Banking('Justine',1200)
print(cust2)

cust1.deposit(500)
print(f'{cust1.cust_name} has {cust1.custBalance} on deposit')

print(f'{cust2.cust_name} has {cust2.custBalance} on deposit')
cust2.deposit(10000)
print(f'{cust2.cust_name} has {cust2.custBalance} on deposit')
print(Banking.bankDeposits)
print(sum(Banking.bankDeposits))
```

```python
# Open an account with Name and an initial deposit amount
# When you print the object have person deposit x amount and now has y amount on deposit

# Open an account
class openAnAccount:
    def __init__(self, bal = 1000):
        self.bal = 10000


    def withdraw(self):
        currentAmount=int(input('Enter the amount to withdraw:'))
        if(currentAmount > self.bal):
            print('Not enough funds to Withdraw!')
        else:
            self.bal = self.bal - currentAmount
            print('Current Balance:' , self.bal)


    def deposit(self):
        currentAmount=int(input('How much are you depositing:'))
        self.bal = self.bal + currentAmount
        print('The Balance you have now is = ',self.bal)
        
    
account= openAnAccount()
account.withdraw()
account.deposit()
```

Without using the built in max function in python, find the maximum value<br>
in the following sequence:<br>
<code>
x = [1,2,0,5,1,4,-1,2]
</code>

```python
x = [1,2,0,5,1,4,-1,2]

def greatNumber(lst):
    theMax = lst[0]
    for num in lst:
        # print(num)
        if theMax < num:
            theMax = num
    return theMax
  
res = greatNumber(x)
print(res)
```
