<h2>Interview Exercise</h2>
Create a class called Numbers<br>
accept 2 numbers<br>
Have a way of adding the numbers<br>
Method to subtract<br>
A method to multiply<br>
Method to divide<br>
print should state the numbers and if they are prime or composite.<br>

```python
import math

class Numbers:

  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2
  
  def __str__(self):
    return f'{self.num1} is : {self.findPrime(self.num1)},  and {self.num2} is : {self.findPrime(self.num2)}'

  def add(self):
    return self.num1 + self.num2

  def subtract(self):
    return self.num1 - self.num2

  def multiply(self):
    return self.num1 * self.num2

  def divide(self):
    return self.num1 / self.num2

  def findPrime(self, num):
    if(num > 1):
      prime = True
      for i in range(2, int(math.sqrt(num) + 1)):
        if num % i == 0:
          prime = False
          break
    return 'Prime' if prime else 'Composite'

nums1 = Numbers(5, 10)
print(f'The result of addition is {nums1.add()}')
print(f'The result of subtraction is {nums1.subtract()}')
print(f'The result of multiplication is {nums1.multiply()}')
print(f'The result of division is {nums1.divide()}')
print(nums1)
```

<h2>self can be replaced with any text</h2>

```python
class Numbers:
    def __init__(shaq,a,b):
        shaq.a = a
        shaq.b = b
        
    def calc_sum(shaq):
        return shaq.a + shaq.b
    
    def __str__(shaq):
        return f'Adding {shaq.a} + {shaq.b} = {shaq.calc_sum()}'
    
n = Numbers(4,7)
print(n)
```

```python
import pandas as pd

df = pd.read_csv('US_Presidents_unordered.txt')
df['Number'] = df["'Name'"].map(lambda x : int(x.replace("'", '').replace('.', '').split()[0]))
df["'Name'"] = df["'Name'"].map(lambda x : ' '.join(x.replace("'", '').replace('.', '').split()[1:]))
df["'Term Start'"] = df["'Term Start'"].map(lambda x : x.replace("'", '').replace('/', '')).astype('int64')
df["'Term End'"] = df["'Term End'"].map(lambda x : x.replace("'", '').replace('/', '')).astype('int64')
df = df.sort_values(by=['Number'])
df.drop(columns = ["Number"], inplace=True)
print(df.dtypes)
df
```
