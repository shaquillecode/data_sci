```python
import pandas as pd
data = { 'New York':[20,15], 
    'California':[10,3], 
    'Rhode Island':[5,6], 
    'Texas': [15,3], 
    'Florida':[8,9], 
    'New Hampshire':[3,4], 
    'New Jersey':[12,8]}
state_info = pd.DataFrame(data.values(),index = data.keys(),columns=['Yes','No']).astype('float32')
state_info['Total votes'] = state_info['Yes'] + state_info['No']
state_info['% Yes'] = ((state_info['Yes']/(state_info['Yes'] + state_info['No']))*100).astype('float64').round(2)
state_info['% No'] = ((state_info['No']/(state_info['Yes'] + state_info['No']))*100).astype('float64').round(2)
state_info.pop('Total votes')
state_info.pop('Yes')
state_info.pop('No')
state_info
# for state in state_info.iterrows():
for state in state_info.itertuples():    
    # print(f'STATE: {state[0]} - {state[1]} - {state[2]}  Type {type(state[0])}')
    # print('='*20)
    df = pd.DataFrame( {state[0]: [state[1],state[2]]}  , index= ['Yes','No'])
    df.plot.pie(y=state[0], figsize=(5, 5))
```

```python

# What percentage of the freshman students drop out before senior year?

# The distribution of students on a college campus was the following
# Freshman: 1500, Sophomore: 1300, Juniors: 1250, Seniors: 1100
# Graph the results as a horizontal bar chart

import pandas as pd
students = [1500, 1300, 1250,1100]
index = ['Freshman', 'Sophomore', 'Juniors','Seniors']
df = pd.DataFrame({'students':students}, index = index)
df.plot.bar()
```

```python
#  Graph the results as a percentage of the student body
import pandas as pd
data = { 'Freshman':[1500], 
    'Sophomore':[1300], 
    'Juniors':[1250], 
    'Seniors': [1100]}
stu_info = pd.DataFrame(data.values(),index = data.keys(),columns=['Total']).astype('float32')
stu_info['Student Body'] = stu_info.Total[0] + stu_info.Total[1] + stu_info.Total[2] + stu_info.Total[3]
stu_info['Freshman %'] = ((stu_info.Total[0]/(stu_info['Student Body']))*100).astype('float64').round(2)
stu_info['Sophmore %'] = ((stu_info.Total[1]/(stu_info['Student Body']))*100).astype('float64').round(2)
stu_info['Junior %'] = ((stu_info.Total[2]/(stu_info['Student Body']))*100).astype('float64').round(2)
stu_info['Senior %'] = ((stu_info.Total[3]/(stu_info['Student Body']))*100).astype('float64').round(2)


stu_info
```

```python
#What percentage of the freshman students drop out before senior year?
df = pd.DataFrame({'val':[1500, 1300, 1250, 1100]}, index = ['Freshman', 'Sophomore', 'Junior','Senior'])
df
df.plot.pie(y='val', figsize = (5,5))
dropout = float(round(((df.loc['Freshman'] - df.loc['Senior'])/df.loc['Freshman'])*100,2))
print(f"The percentage of Freshman who drop out before senior year is %{dropout}")
```
