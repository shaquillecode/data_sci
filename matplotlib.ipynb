<h2>Step 1 - Find the top 3 for 2015</h2>

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'data/Bias_crime_2015_2019.csv')
df2 = df[['Race_Ethnicity','2015']].sort_values(by=['2015'], ascending=False)[:3]
df2 = df[['Race_Ethnicity','2015']].sort_values(by=['2015'], ascending=False)[:3]
ethn3 = df2['Race_Ethnicity'].to_list()
print(ethn3)
```

```python
df3 = df[df.Race_Ethnicity.isin(ethn3)]
df3
data = df[df.Race_Ethnicity.isin(ethn3)].iloc[:, 1:]
data
allData = []
for line in data.itertuples():
    # print(line[1],line[2],line[3],line[4],line[5])
    allData.append([line[1],line[2],line[3],line[4],line[5]])
allData
X = np.arange(2015, 2020)

plt.bar(X + 0.00, allData[0], color="b", width=0.20)
plt.bar(X + 0.25, allData[1], color="g", width=0.20)
plt.bar(X + 0.50, allData[2], color="r", width=0.20)

plt.show()
```
