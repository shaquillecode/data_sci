```python
import pandas as pd

desserts ={'Apple Pie': 9.99,
          'Cheesecake': 21.00, 
          'Chocolate Souffle': 14.99,
          'Carrot Cake': 11.99}
print(desserts.keys())
print(desserts.values())
results = pd.Series(desserts.values(), index = desserts.keys())
results.loc['Cheesecake']


sum = round(results.sum(), 2)
print(f'the sum is: {sum}')

average = round(results.mean(), 2)
print(f'the average is: {average}')
```

<h2>Pandas Series - Basic Statistics</h2>
To gather basic statistics on a data Series we can use the built in<br>
'.describe' method <br>
<code>
import pandas as pd
sales = {'January': 150, 'February':200, 'March': 175,
        'April': 75, 'May': 80, 'June': 300,
        'July': 250, 'August':130, 'September': 195,
        'October': 75, 'November': 120, 'December': 400}
ps = pd.Series([int(val)+50 for val in sales.values()],
               index = [mon.upper()[:3] for mon in sales.keys()])
ps.describe()
</code><br><br>
This will give us the following statistics:<br>
1 - count<br>
2 - mean<br>
3 - standard deviation<br>
4 - minimum<br>
5 - 25% quartile<br>
6 - Median<br>
7 - 75% quartile<br>
5 - maximum<br>

```python
import pandas as pd
sales = {'January': 150, 'February':200, 'March': 175,
        'April': 75, 'May': 80, 'June': 300,
        'July': 250, 'August':130, 'September': 195,
        'October': 75, 'November': 120, 'December': 400}
ps = pd.Series([int(val)+50 for val in sales.values()],
               index = [mon.upper()[:3] for mon in sales.keys()])
ps.describe()
```

```python
import pandas as pd
sales = {'January': 150, 'February':200, 'March': 175,
        'April': 75, 'May': 80, 'June': 300,
        'July': 250, 'August':130, 'September': 195,
        'October': 75, 'November': 120, 'December': 400}
ps = pd.Series([int(val)+50 for val in sales.values()],
               index = [mon.upper()[:3] for mon in sales.keys()])
ps.count()
ps.mean()
ps.std()
ps.min()
ps.max()
ps.median()
```

The following represented the price of gasoline in the following states:<br>

Create a pandas series with the index of the state and the column with the average price<br>

```python
gas_prices = {'New York':[3.23,3.17,3.23,3.10,3.09],
              'Connecticut':[3.15,3.23,3.11,3.15,3.09]}

res = pd.Series([sum(v)/len(v) for v in gas_prices.values()], index = [k for k in gas_prices.keys()]).sort_values(ascending=False)

print(res)
```

We can also sort a series using a lambda expression.  Consider the <br>

<code>

ps.sort_values(key=lambda x: x.str.lower())
ps

NOTE: We can use the str method to change all values in a series to a string

```python
import pandas as pd
ps = pd.Series(['a', 'c', 'D', 'e',  'B'])
ps = ps.str.lower()
ps.sort_values(inplace = True)
ps
```
