```python
import pandas as pd
df = pd.read_csv(r'baseball.csv')
teams = df['team'].unique()
years = df['year'].unique()
# teams
for team in teams:
    for year in years:
        # print(year,team)
        df_max = df[['player','year','rbi','team']][df.team == team][df.year == year]['rbi'].max()
        if df_max:
            df_player = df[['player','year','rbi','team']][df.team == team][df.year == year][df.rbi == df_max]
            # if df_player:
            print(df_player)
# df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bb = pd.read_csv(r'baseball.csv')
# 4 - Create a chart of all h/ab > .250 as below average vs > .250 above average for each year
bb["batting_ave"] = bb["h"]/bb["ab"]

bb["batting_ave"].mean().round(4) # 0.2013

yr = bb['year'].unique()

for y in yr:
    res = bb[bb.year == y]["batting_ave"]
    abv = res[res >= 0.25].count()
    blw = res[res < 0.25].count()
    # PIE CG+HART
    baChart = pd.DataFrame({"BA": [blw, abv]}, index=["below_ave", "above_ave"]).plot.pie(subplots=True, figsize=(5,5))
    plt.title(f"{y} BA chart (ave. == 0.250)", fontdict={"color":"white", "size": 20})
    # HORIZONTAL BAR CHART
    baChart = pd.DataFrame({"BA": [blw, abv]}, index=["below_ave", "above_ave"]).plot.barh()
    plt.title(f"{y} BA chart (ave. == 0.250)", fontdict={"color":"black", "size": 20})
```
